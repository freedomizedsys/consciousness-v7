```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import seaborn as sns

plt.style.use('seaborn-v0_8')
%matplotlib inline
```

## 核心動態系統定義 / Core Dynamic System Definition

```python
def consciousness_system(t, y, params):
    C, M, P = y
    k, v, alpha, gamma, lambda_mem, W, K, Gp, R, f, H, beta = params
    
    dC_dt = M * (W * K * Gp / R) * np.exp(-gamma * P) * (1 + alpha * M) * v
    dM_dt = lambda_mem * C
    dP_dt = -k * P
    
    return [dC_dt, dM_dt, dP_dt]
```

## 四種情境模擬與圖表生成 / Four Scenarios Simulation and Figure Generation

```python
def run_scenario(k, P0, R, label_en, label_zh):
    params = [k, 0.05, 0.2, 0.05, 0.01, 1.0, 5.0, 0.8, R, 0.5, 1.0, 1.0]
    sol = solve_ivp(consciousness_system, [0, 100], [0.1, 0.0, P0], args=(params,), method='RK45', rtol=1e-6, atol=1e-8)
    t = sol.t
    C = sol.y[0]
    M = sol.y[1]
    return t, C, M

# 四種情境 / Four Scenarios
scenarios = [
    (0.8, 8.0, 1.0, 'Sc1: High Pain/High Trans (k=0.8)', 'Sc1: 高痛苦高轉化 (k=0.8)'),
    (0.2, 1.0, 1.0, 'Sc2: Low Pain/High Happiness (k=0.2)', 'Sc2: 低痛苦高快樂 (k=0.2)'),
    (0.5, 4.0, 1.5, 'Sc3: Balanced (k=0.5)', 'Sc3: 平衡情境 (k=0.5)'),
    (0.15, 4.0, 3.0, 'Sc4: High Resistance/Low Trans (k=0.15)', 'Sc4: 高阻力低轉化 (k=0.15)')
]

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.ravel()

for i, (k, P0, R, label_en, label_zh) in enumerate(scenarios):
    t, C, M = run_scenario(k, P0, R, label_en, label_zh)
    
    # 圖 1: C(t) / Figure 1: C(t)
    axs[0].plot(t, C, label=label_zh)
    # 圖 2: M(t) / Figure 2: M(t)
    axs[1].plot(t, M, label=label_zh)
    # 圖 3: F(t) 簡化示意 / Figure 3: F(t) simplified
    F = 1.0 * C * (1 - np.exp(-k * (8.0 - np.minimum(8.0, P0 * np.exp(-0.05 * t)))))
    axs[2].plot(t, F, label=label_zh)

axs[0].set_title('圖 1：意識強度 C(t) 的時間演化曲線\nFigure 1: Time Evolution of Consciousness Intensity C(t)')
axs[1].set_title('圖 2：長期記憶強度 M(t) 的累積曲線\nFigure 2: Memory Accumulation M(t)')
axs[2].set_title('圖 3：快樂 F(t) 的自然湧現曲線\nFigure 3: Natural Emergence of Happiness F(t)')
axs[3].set_title('圖 4：敏感性熱圖（概念示意）\nFigure 4: Sensitivity Heatmap (Conceptual)')

for ax in axs[:3]:
    ax.legend()
    ax.set_xlabel('時間 t / Time t')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
